```python
def print_commit(repo, path, x):
    print(f'df.loc[("{repo}", "{path}"), "date"] = iso("{x.commit.committer.date}")')
    print(" ", x.commit.message)
    print(" ", x.commit.html_url)
    print()


def get_history(repo, path, *, filter=True):
    record = []
    found = False
    for x in g.get_repo(repo).get_commits(path=path):
        record.append(x)
        msg = x.commit.message
        if (
            not filter
            or "517" in msg
            or "skbuild" in msg
            or "scikit-build" in msg.replace("_", "-")
        ):
            print_commit(repo, path, x)
            found = True

    if filter and not found:
        print("No filtered commits found, printing all commits\n")
        for x in record:
            print_commit(repo, path, x)
```
